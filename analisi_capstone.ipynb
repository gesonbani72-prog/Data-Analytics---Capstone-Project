## Download dati ISTAT

```python
# Importo le librerie utili per scaricare i dati tramite API
import requests
import pandas as pd
import io
```

```python
# Inizio a creare la variabile URL con url_istat e nella request uso headers inserendo il suggerimento dato nel file d'esame
url_istat = 'https://esploradati.istat.it/SDMXWS/rest/data/41_983' # url preso dal file d'esame
dati_istat = requests.get(url_istat, headers={'Accept': 'application/vnd.sdmx.data+csv;version=1.0.0'}) #headers preso dal file d'esame
```

```python
print(f'1)lo stato codice è {dati_istat.status_code}')
print(f'2)il formato ottenuto è {type(dati_istat)}')
print(f'3)verifico velocemente se sono presenti dei dati:{dati_istat.text[:200]}') #nelle lezioni abbiamo usato .json() ma in questo caso non ha funzionato essendo formato diverso e ho quindi usato .text
```

```python
df_istat = pd.read_csv(io.StringIO(dati_istat.text), sep=",")
df_istat
```
