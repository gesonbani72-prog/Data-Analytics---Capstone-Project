## Download dati ISTAT

```python
# Importo le librerie utili per scaricare i dati tramite API
import requests
import pandas as pd
```

```python
# Inizio a creare la variabile URL con url_istat e nella request uso headers inserendo il suggerimento dato nel file d'esame
url_istat = 'https://esploradati.istat.it/SDMXWS/rest/data/41_983' # url preso dal file d'esame
dati_istat = requests.get(url_istat, headers={'Accept': 'application/vnd.sdmx.data+csv;version=1.0.0'}) #headers preso dal file d'esame
print(f'lo stato codice è {dati_istat.status_code}')
```
